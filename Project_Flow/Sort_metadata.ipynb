```python
for participant_id, participant_data in HMP.groupby('Participant ID'):
    for column in HMP.columns:
        if HMP[column].dtypes == 'object' :
            value = participant_data[column].dropna().unique()
            if len(value) == 1:
                HMP.loc[HMP['Participant ID'] == participant_id, column] = value[0]
            else:
                if len(value) > 0:
                    print(f'no val{participant_id}\t{value}')

# HMP.loc[HMP['Participant ID'] == 'C3005',:]
```

```python
nonIBD = HMP.loc[HMP['diagnosis'] == 'nonIBD']
UC = HMP.loc[HMP['diagnosis'] == 'UC']
CD = HMP.loc[HMP['diagnosis'] == 'CD']
```

```python
CD.fillna(CD.select_dtypes(include='object').mode().iloc[0], inplace=True)
for col in CD.columns:
    if CD[col].dtypes != 'object' and CD[col].sum() > 0: #and col not in ['Age at diagnosis','General wellbeing']:
            CD[col] = np.random.uniform(CD[col].min(),CD[col].max(), len(CD))
```

```python
CD.week_num = round(CD.week_num)
CD.interval_days = round(CD.interval_days)
CD.visit_num = round(CD.visit_num)
CD['Age at diagnosis'] = round(CD['Age at diagnosis'])
CD.consent_age = round(CD.consent_age)
CD.Weight = round(CD.Weight)
CD['Weight.1'] = round(CD['Weight.1'])
CD.Height = round(CD.Height)
CD["Modified Baron's Score"] = round(CD["Modified Baron's Score"])
CD['Total.1'] = round(CD['Total.1'])
CD['Total'] = round(CD['Total'])
CD['Total.2'] = round(CD['Total.2'])
CD['Total.3'] = round(CD['Total.3'])
CD['SIBDQ Score'] = round(CD['SIBDQ Score'])
CD['sccai'] = round(CD['sccai'])
CD['SES-CD Score'] = round(CD['SES-CD Score'])

CD
```

```python
for col in CD.columns:
    if col in ['External ID','Participant ID','diagnosis','data_type']:
        continue
    i = rd.uniform(0,0.3)
    CD.loc[CD.sample(frac=i).index, col] = np.nan
CD.to_csv('~/Desktop/CD.csv')
```
