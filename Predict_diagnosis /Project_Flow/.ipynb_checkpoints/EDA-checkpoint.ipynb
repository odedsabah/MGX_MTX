```python
Target = METADATA.diagnosis
METADATA_Numeric = METADATA.select_dtypes(include = ['float64', 'int64'])
METADATA_Categories = METADATA.select_dtypes(include = [object]).astype("category")
METADATA_Categories = METADATA_Categories.loc[:,METADATA_Categories.columns != 'date_of_receipt']
METADATA_Numeric
```

```python
def Check_distribution(METADATA_Numeric):
    df = pd.DataFrame()
    feat = []
    Distribution = []
    for column in METADATA_Numeric.columns:
        feat.append(column)
        stat, p_val  = stats.shapiro(METADATA_Numeric.dropna()[feat])
        if p_val >= 0.05:
            Distribution.append('Normally')
        else:
            Distribution.append('Asymmetrically')
    df['Feature'] = feat
    df['Distribution'] = Distribution
    return df

Table_Distribution = Check_distribution(METADATA_Numeric)
display(Table_Distribution)
Table_Distribution.Distribution.value_counts()
```

```python
def Comparing_Variables(METADATA_Numeric):
    df = pd.DataFrame()
    feat1s = []
    feat2s = []
    p_values = []
    for feat1 in METADATA_Numeric.columns:
        for feat2 in METADATA_Numeric.columns:
            if feat1 != 'week_num' and feat2 != 'interval_days' and feat1 != feat2:
                feat1s.append(feat1)
                feat2s.append(feat2)
                cor_coef, Pvalue = stats.wilcoxon(METADATA_Numeric[feat1], METADATA_Numeric[feat2],nan_policy='omit',alternative= 'two-sided', zero_method= 'zsplit' )
                p_values.append(Pvalue)
            if feat1 == 'week_num' and feat2 == 'interval_days':
                feat1s.append(feat1)
                feat2s.append(feat2)
                stats_ttest = stats.ttest_rel(METADATA_Numeric.dropna()[feat1], METADATA_Numeric.dropna()[feat2],alternative='two-sided', nan_policy='omit')
                p_values.append(stats_ttest.pvalue)
    df['Feature_1'] = feat1s
    df['Feature_2'] = feat2s
    df['p_value'] = p_values
    return df
CC = Comparing_Variables(METADATA_Numeric.iloc[1:30,1:30])
CC
```

```python
for feat in METADATA_Numeric:
    sns.displot(data=METADATA, x=METADATA_Numeric[feat], hue=Target, kind='kde')
    plt.show()
```

```python
def Crames_v_Chi(METADATA_Categories):
    df = pd.DataFrame()
    feat1s = []
    feat2s = []
    corrs = []
    p_values = []
    for feat1 in METADATA_Categories.columns:
        for feat2 in METADATA_Categories.columns:
            if feat1 != feat2:
                feat1s.append(feat1)
                feat2s.append(feat2)
                chi2, p, dof, ex = chi2_contingency(pd.crosstab(METADATA_Categories[feat1], METADATA_Categories[feat2]),correction=True)
                corrs.append(chi2)
                p_values.append(p)

    df['Feature_1'] = feat1s
    df['Feature_2'] = feat2s
    df['Correlation'] = corrs
    df['p_value'] = p_values
    display(df)
Crames_v_Chi(METADATA_Categories.iloc[10:30,10:30])
```
